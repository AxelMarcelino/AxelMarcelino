# Agglomerative clustering untuk dataset random

### Library yang diperlukan

```python
# import library yang diperlukan
import numpy as np
import pandas as pd
from scipy import ndimage
from scipy.cluster import hierarchy
from scipy.spatial import distance_matrix
from matplotlib import pyplot as plt
from sklearn import manifold, datasets
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
%matplotlib inline
```

## Agglomerative Clustering Single Linkage

### Generate random data

```python
# Melakukan generate random data
# untuk menghasilkan dataset kita menggunakan make_blobs
# untuk parameternya:
    # n_sampels = jika int, merupakan jumlah poin yang dibagi rata" di antara cluster. Jika array, elemen urutan menunjukkan jumlah sampel per cluster
    # centers = jumlah pusat yang akan dihasilkan, atau lokasi pusat tetap
    # cluster_std = standard deviasi dari cluster

x1, y1 = make_blobs(n_samples=50, centers=[[4,4], [-2,-1], [1,1], [10,4]], cluster_std=0.9, random_state=101)
```

### Plotting random data

```python
# melakukan ploting terhadap x1 dengan marker "o"

plt.scatter(x1[:, 0], x1[:, 1], marker="o")
```

### Agglomerative Clustering

```python
# melakukan clustering dengan agglomerative clustering
# agglomerative clustering membutuhkan 2 paramter:
    # n_clusters = jumlah cluster yang akan dibentuk serta jumlah sentroid yang akan dibentuk
    # linkage = kriteria linkage mana yang akan digunakan (complete, average, single)

agglom_single = AgglomerativeClustering(n_clusters=4, linkage="single")
agglom_single.fit(x1, y1)
```

```python
# membuat figure dengan ukuran 12 inc dan 6 inc
# dengan paramter figsize, (float, float) yang merepresentasikan lebar dan tinggi dalam inch

plt.figure(figsize=(12, 6))

# mencari nilai minimum dan maximum dari rentang x1
x_min, x_max = np.min(x1, axis=0), np.max(x1, axis=0)


# mencari rata-rata dari jarax x1
x1 = (x1 - x_min) / (x_max - x_min)

# loop untuk menampilkan semua datapoints
for i in range(x1.shape[0]):
    # ganti datapoint dengan nilai cluster respectivenya
    # (cth. 0) dan waranya dengan colormap (plt.cm.spectral)
    plt.text(x1[i, 0], x1[i, 1], str(y1[i]),
             color=plt.cm.nipy_spectral(agglom_single.labels_[i] / 10),
             fontdict={"weight": "bold", "size": 12})

# hapus ticks x, ticks y, x dan y axis
plt.xticks([])
plt.yticks([])

# menampilkan plot dari data original sebelum clustering
plt.scatter(x1[:, 0], x1[:, 1], marker=".")
# menampilkan plot
plt.show
```

Single Linkage, prosedur ini didasarkan pada jarak terkecil. Jika dua obyek terpisah oleh jarak yang pendek maka kedua obyek tersebut akan digabung menjadi satu cluster daan demikian saterusnya.

## Data Cleaning

```python
# menampilkan shape dari dataset sebelum data cleaning
print("Shape of dataset before cleaning: ", pdf.size)

# mengubah fitur-fitur yang dipilih menjadi numeric
# bertujuan agar fitur yang memiliki nilai null ($null$) menjadi NaN
pdf[["sales", "resale", "type", "price", "engine_s",
     "horsepow", "wheelbas", "width", "length", "curb_wgt", "fuel_cap",
     "mpg", 'lnsales']] = pdf[["sales", "resale", "type", "price", "engine_s",
     "horsepow", "wheelbas", "width", "length", "curb_wgt", "fuel_cap",
     "mpg", 'lnsales']].apply(pd.to_numeric, errors = "coerce")

# membuang baris yang memiliki nilai NaN (missing value) dengan cara didrop
pdf = pdf.dropna()

# mereset index dari dataframe, dan menggunakan default sebagai gantinya
# drop=True berguna untuk tidak membasukkan "index" ke kolom dataframe
pdf = pdf.reset_index(drop=True)

# menampilkan shape dari dataset setelah data cleaning
print("Shape of dataset after cleaning: ", pdf.size)
pdf.head()
```

## Feature Selection

```python
# memilih fitur yang akan digunakan untuk melakukan clustering
featureset = pdf[["engine_s", "horsepow", "wheelbas", "width", "length", "curb_wgt", "fuel_cap", "mpg"]]
featureset.head()
```

## Normalization

```python
# melakukan normalisasi data agar berbedaan nilai data tidak jauh
# sehingga tidak ada data yang memiliki impact lebih dalam clustering

from sklearn.preprocessing import MinMaxScaler

# mengubah data fitur menjadi bentuk numpy array
x = featureset.values

# melakukan normalisasi
min_max_scaler = MinMaxScaler()
feature_mtx = min_max_scaler.fit_transform(x)
feature_mtx[0:5]
```

## Clustering dengan Scipy

```python
import scipy

# mengambil banyak baris dalam dataset yang telah dinormalisasi
length = feature_mtx.shape[0]

# membuat matrix dengan nilai 0 dengan ukuran (length x length)
D = scipy.zeros([length, length])

# mencari jarak setiap data dan disimpan dalam matrix D
for i in range(length):
    for j in range(length):
        D[i, j] = scipy.spatial.distance.euclidean(feature_mtx[i], feature_mtx[j])
```

### Single Linkage

```python
import pylab
import scipy.cluster.hierarchy
z = hierarchy.linkage(D, "single")
```

```python
from scipy.cluster.hierarchy import fcluster
max_d = 3
clustersz = fcluster(z, max_d, criterion="distance")
clustersz
```

```python
from scipy.cluster.hierarchy import fcluster
k=5
clustersz = fcluster(z, k, criterion="maxclust")
clustersz
```

### Plotting Dendrogram

```python
# membuat figure dengan ukuran 20 inc dan 50 inc
# dengan paramter figsize, (float, float) yang merepresentasikan lebar dan tinggi dalam inch
fig = plt.figure(figsize=(20, 70))

def llf(id):
    return "[%s %s %s]" % (pdf["manufact"][id], pdf["model"][id], int(float(pdf["type"][id])))

dendro = hierarchy.dendrogram(z, leaf_label_func=llf, leaf_rotation=0, leaf_font_size=12, orientation="right")
```

### Average Linkage

```python
import pylab
import scipy.cluster.hierarchy
y = hierarchy.linkage(D, "average")
```

```python
from scipy.cluster.hierarchy import fcluster
max_d = 3
clustersy = fcluster(y, max_d, criterion="distance")
clustersy
```

```python
from scipy.cluster.hierarchy import fcluster
k=5
clustersy = fcluster(y, k, criterion="maxclust")
clustersy
```

### Ploting dendrogram

```python
# membuat figure dengan ukuran 20 inc dan 50 inc
# dengan paramter figsize, (float, float) yang merepresentasikan lebar dan tinggi dalam inch
fig = plt.figure(figsize=(10, 50))

# membuat fungsi untuk menampilkan nama dari leaf(daun) / nama manufaktur mobil + model + type
def llf(id):
    return "[%s %s %s]" % (pdf["manufact"][id], pdf["model"][id], int(float(pdf["type"][id])))

dendro = hierarchy.dendrogram(y, leaf_label_func=llf, leaf_rotation=0, leaf_font_size=12, orientation="right")
```

```python
import matplotlib.cm as cm

n_clusters = max(sl_agglom_complete.labels_) + 1
colors = cm.rainbow(np.linspace(0, 1, n_clusters))
cluster_labels = list(range(0, n_clusters))

# membuat figure dengan ukuran 12x12 inch
plt.figure(figsize=(12, 12))

for color, label in zip(colors, cluster_labels):
    subset = pdf_copy3[pdf_copy3.cluster_ == label]

    for i in subset.index:
        plt.text(subset.sepal_length[i], subset.petal_length[i], str(subset['species'][i]), rotation=25)
    plt.scatter(subset.sepal_length, subset.petal_length, c=color, label='cluster'+str(label), alpha=.5)

plt.legend()
plt.title('Clusters')
plt.xlabel('sepal-l')
plt.ylabel('petal-l')
```

```python
# Menghitung jumlah kasus di setiap kelompok
pdf_copy3.groupby(['cluster_','species'])['cluster_'].count()
```

```python
# Melihat karakteristik dari masing-masing cluster
agg_iris3 = pdf_copy3.groupby(['cluster_','species'])['sepal_length','sepal_width','petal_length','petal_width'].mean()
agg_iris3
```

```python
# Plotting Dendogram
plt.figure(figsize=(12,10))
# plotting dendrogram
for color, label in zip(colors, cluster_labels):
  subset = agg_iris3.loc[(label,),]
  for i in subset.index:
    plt.text(subset.loc[i][0], subset.loc[i][2], str(i))

  plt.scatter(subset.sepal_length, subset.petal_length, c=color, label='cluster'+str(label))

plt.legend()
plt.title('Clusters')
plt.xlabel('sepallength')
plt.ylabel('petallength')
```
