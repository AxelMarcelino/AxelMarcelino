# **Mengimpor Library**

```python
import pandas as pd
from numpy import unique
from numpy import where
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
from sklearn.cluster import KMeans
%matplotlib inline
```

Memuat berbagai Library yang dipakai dan membuat matplotlib sebagai *inline*, serta jenis graph matplotlib plot sebagai 'seaborn-whitegrid'.

# **Mengimpor Dataset**

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris.data
```

Memuat dataset iris untuk melakukan clustering dan menampilkan data dataset iris.

# **Original Cluster  (fitur 1 dan 2)**



```python
features1 = iris.data.T
plt.plot()
plt.scatter(features1[1], features1[2])
plt.show()
```

Menampilkan pola Original Cluster dataset Iris dengan fitur 1 dan 2.

# **Elbow Method  (fitur 1 dan 2)**



```python
X = np.array(list(zip(features1[1], features1[2]))).reshape(len(features1[2]), 2)
SSE = []
K = range(1,10)
for k in K:
  kmeanModel = KMeans(n_clusters=k).fit(X)
  kmeanModel.fit(X)
  SSE.append(kmeanModel.inertia_)

plt.plot(K, SSE, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

Menampilkan Elbow Method dataset iris nilai K terbaik fitur 1 dan 2, yang di lihat dari siku sebagai titik utama yaitu K=2.

# **Original Cluster (fitur 2 dan 3)**



```python
features2 = iris.data.T
plt.plot()
plt.scatter(features2[2], features2[3])
plt.show()
```

Menampilkan pola Original Cluster dataset Iris dengan fitur 2 dan 3.

# **Elbow Method (fitur 2 dan 3)**


```python
X1 = np.array(list(zip(features2[2], features2[3]))).reshape(len(features2[3]), 2)
SSE1 = []
K1 = range(1,10)
for k1 in K1:
  kmeanModel1 = KMeans(n_clusters=k1).fit(X1)
  kmeanModel1.fit(X1)
  SSE1.append(kmeanModel1.inertia_)

plt.plot(K1, SSE1, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

Menampilkan Elbow Method dataset iris nilai K terbaik fitur 2 dan 3, yang di lihat dari siku sebagai titik utama yaitu K=2.

# **Evaluasi (fitur 1 dan 2)**


```python
for k in K:
  kmeanModel = KMeans(n_clusters=k, random_state=1).fit(X)
  kmeanModel.fit(X)
  labels = kmeanModel.predict(X)
  interia = kmeanModel.inertia_
  print("k:", k, " cost:", interia)
print("")
```

Hasil evaluasi dataset iris nilai K terbaik fitur 1 dan 2 dataset iris yaitu K=2 dengan cost: 86.31. Pada K=1 - K=2 penurunan nilai K masih relevan, sedangkan dari K=3 - K=9 penurunan nilai K nya sudah tidak relevan lagi. Maka batas nilai k sudah sesuai pada Elbow method ini.

```python
from sklearn.metrics.cluster import silhouette_score
silhouette_score(X, labels)
```

Nilai Silhouette Coefficient dataset iris yaitu 0,39194 yang mendekati 1 dan sudah efektif pada pengklasifikasian cluster fitur 1 dan 2.  

# Evaluasi (fitur 2 dan 3)


```python
for k1 in K1:
  kmeanModel1 = KMeans(n_clusters=k1, random_state=1).fit(X1)
  kmeanModel1.fit(X1)
  labels1 = kmeanModel1.predict(X1)
  interia1 = kmeanModel1.inertia_
  print("k:", k1, " cost:", interia1)
print("")
```

Hasil evaluasi dataset iris nilai K terbaik fitur 2 dan 3 dataset iris yaitu K=2 dengan cost: 86.39. Pada K=1 - K=2 penurunan nilai K masih relevan, sedangkan dari K=3 - K=9 penurunan nilai K nya sudah tidak relevan lagi. Maka batas nilai k sudah sesuai pada Elbow method ini.

```python
from sklearn.metrics.cluster import silhouette_score
silhouette_score(X1, labels1)
```

Nilai Silhouette Coefficient dataset iris yaitu 0,58508 yang mendekati 1 dan sudah efektif pada pengklasifikasian cluster fitur 2 dan 3.

# **Visualisasi Clustering K=2 (fitur 1 dan 2)**


```python
y_pred = KMeans(n_clusters=2).fit_predict(X)
plt.plot
LABEL_COLOR_MAP = {0 : 'r',
                   1 : 'g',
                   2 : 'b'
                   }

label_color = [LABEL_COLOR_MAP[l] for l in y_pred]
plt.scatter(features1[1], features1[2], c=label_color)
plt.title("K=2")
plt.show()
```

Tampilan visualisai Clustering pada dataset iris dengan K=2.
